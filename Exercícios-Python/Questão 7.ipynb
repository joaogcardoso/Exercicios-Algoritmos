```python
try:
    lista = []
    quantidade = int(input("Digite a quantidade de valores a serem inseridos:"))
    def funcaoSoma(quantidade):
        for i in range(0, quantidade):
            numero = int(input(f"Digite o {i + 1}º:"))
            lista.append(numero)
        print(f"A soma da lista é igual a {sum(lista)}")
    funcaoSoma(quantidade)
except ValueError:
    print("Entrada Inválida!")
```
