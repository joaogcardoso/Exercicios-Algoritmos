```python
try:
    lista = []
    quantidade = int(input("Digite a quantidade de números a serem inseridos:"))
    def funcaoMedia(quantidade):
        for i in range(0, quantidade):
            numero = int(input(f"Digite o {i + 1}º número:"))
            lista.append(numero)
        print(f"A média dos valores da lista é igual a {sum(lista) / quantidade}")
    funcaoMedia(quantidade)
except ValueError:
    print("Entrada Inváldia!")
```
