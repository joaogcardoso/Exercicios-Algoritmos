```python
try:
    n1 = int(input("Digite o 1º número: "))
    n2 = int(input("Digite o 2º número: "))

    dic_resultados = {
    "Soma:": n1 + n2,
    "Subtração:": n1 - n2,
    "Multiplicação:": n1 * n2,
    "Divisão:": n1 / n2
}

except ValueError:
    print("Entrada inválida!")

def soma(n1, n2):
    dic_resultados = {n1 + n2} 
    return dic_resultados
def subtracao(n1, n2):
    dic_resultados = {n1 - n2}
    return dic_resultados
def multiplicacao(n1, n2):
    dic_resultados = {n1 * n2}
    return dic_resultados
def divisao(n1, n2):
    if n2 == 0:
        print("Erro! Divisão por zero!")
    else:
        dic_resultados = {n1 / n2}
        return dic_resultados

for operacao, resultado in dic_resultados.items():
    print(operacao, resultado)
```
