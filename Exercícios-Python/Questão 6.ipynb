```python
try:
    numero = int(input("Digite um número inteiro:"))
    def numero_impar(numero):
        print(f"Números ímpares até {numero}:")
        for i in range(0, numero + 1):
            if i % 2 != 0:
                print(i)
    numero_impar(numero)
except ValueError:
    print("Entrada invalida!")
```
