```python
n1 = float(input("Digite o 1º número:"))
n2 = float(input("Digite 2º um número:"))
n3 = float(input("Digite 3º um número:"))

media = (n1 + n2 + n3) / 3
mediaFinal = round(media, 1)

print("Média:", mediaFinal)

if media >= 7 :
    print("Aprovado")
elif media < 5 :
    print("Reprovado")
else :
    print("Recuperação")
```
