```python
import random

numero = 0

min = int(input("Digite o valor mínimo para o número secreto:"))
max = int(input("Digite o valor máximo para o número secreto::"))

numeroSecreto = random.randint(min, max)

print("O número secreto está no intervalo entre", min, "e", max, "\n")
while numeroSecreto != numero :
    numero = int(input("Digite um número:"))

    if numero == numeroSecreto :
        print("Acertou! O número secreto era", numeroSecreto)
    else :
        print("Errou! Tente novamente")
```
