```python
maiorNumero = 0

n1 = int(input("Digite o 1º número:"))
n2 = int(input("Digite 2º um número:"))
n3 = int(input("Digite 3º um número:"))

if n1 > maiorNumero or n1 < maiorNumero :
    maiorNumero = n1
if n2 > maiorNumero :
    maiorNumero = n2
if n3 > maiorNumero :
    maiorNumero = n3
print("O maior número é:", maiorNumero)
```
